```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator  # 

plt.rcParams["font.family"] = ["sans-serif"]
plt.rcParams["axes.unicode_minus"] = False  # 

def plot_train_loss(csv_files, labels, x_label='epoch', y_label='train_loss', 
                    title='Training Loss Comparison', figsize=(10, 6)):
    plt.figure(figsize=figsize)
    
    # CSV
    for file, label in zip(csv_files, labels):
        try:
            df = pd.read_csv(file)
            
            if x_label not in df.columns:
                raise ValueError(f" {file}  {x_label} ")
            if y_label not in df.columns:
                raise ValueError(f" {file}  {y_label} ")
            
            # 
            plt.plot(df[x_label], df[y_label], label=label)
            
        except Exception as e:
            print(f" {file} : {str(e)}")
    
    # 
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.legend()
    plt.grid(alpha=0.3)
    
    # x
    ax = plt.gca()  # 
    ax.xaxis.set_major_locator(MaxNLocator(integer=True))  # 
    
    plt.tight_layout()
    plt.show()

# 
if __name__ == "__main__":
    csv_file_paths = [
        "runs/vitb16_text_only/logs/2025-11-10_16-02-13.csv",
        "runs/vitb16_linear_probe/logs/2025-11-10_16-08-21.csv",
        "runs/vitb16_lora_only/logs/2025-11-10_16-13-30.csv",
        "runs/vitb16_full/logs/2025-11-10_16-22-40.csv"
    ]
    
    model_labels = [
        "text-only",
        "linear-probe",
        "lora-only",
        "full"
    ]
    
    plot_train_loss(
        csv_files=csv_file_paths,
        labels=model_labels,
        title="Image-to-Text Recall@1 of model"
    )
```
